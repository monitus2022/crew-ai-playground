```python
from crewai import LLM, Agent, Crew, Process, Task
from crewai_tools import (
    DirectoryReadTool,
    FileReadTool
    # WebsiteSearchTool
)
import os
from dotenv import load_dotenv
import yaml

load_dotenv()
```

```python
# Setup LLM from OpenRouter

OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")

llm = LLM(
    model="openrouter/openai/gpt-oss-120b:free",
    base_url="https://openrouter.ai/api/v1",
    api_key=OPENROUTER_API_KEY,
    max_tokens=16000
)
```

```python
# List of Tools

directory_tool = DirectoryReadTool()
file_tool = FileReadTool()
```

```python
# Setup config for agents and tasks

agents_config = "agents.yaml"
with open(agents_config, "r", encoding="utf-8") as f:
    agents_config = yaml.safe_load(f)

tasks_config = "tasks.yaml"
with open(tasks_config, "r", encoding="utf-8") as f:
    tasks_config = yaml.safe_load(f)
```

```python
# List of Agents

coder_agent = Agent(
    config=agents_config["coder"],
    verbose=True,
    tools=[file_tool],
    llm=llm
)
```

```python
# List of Tasks

task = Task(
    config=tasks_config["create_webapp"],
    agent=coder_agent
)
```

```python
# Init Crew and run Process

crew = Crew(
    agents=[coder_agent],
    tasks=[task],
    process=Process.sequential,
    verbose=True,
    tracing=True
)
```

```python
# Testing Crew

user_request = """
    Create a simple single-file web app that displays 'Hello, World!' 
    and has a button to toggle dark mode.
"""

result = crew.kickoff(
    inputs={"user_request": user_request}
)
```

```python
def print_token_usage(result, crew=None, prompt_cost_per_million=0.2, completion_cost_per_million=1.5):
    """Pretty print token usage after crew finishes"""
    print("\n" + "="*70)
    print("CREW FINISHED - TOKEN USAGE REPORT")
    print("="*70)
    
    tu = result.token_usage
    
    print(f"Total Tokens       : {getattr(tu, 'total_tokens', 'N/A'):,}")
    print(f"Prompt Tokens      : {getattr(tu, 'prompt_tokens', 'N/A'):,}")
    print(f"Completion Tokens  : {getattr(tu, 'completion_tokens', 'N/A'):,}")
    print(f"Successful Requests: {getattr(tu, 'successful_requests', 'N/A')}")
    
    # Rough cost estimation
    if hasattr(tu, 'total_tokens'):
        prompt_cost = (getattr(tu, 'prompt_tokens', 0) / 1_000_000) * prompt_cost_per_million
        completion_cost = (getattr(tu, 'completion_tokens', 0) / 1_000_000) * completion_cost_per_million
        print(f"\nEstimated Prompt Cost     : ~${prompt_cost:.4f} USD")
        print(f"Estimated Completion Cost : ~${completion_cost:.4f} USD")
        print(f"Estimated Total Cost     : ~${prompt_cost + completion_cost:.4f} USD")
    
    if crew and hasattr(crew, 'usage_metrics'):
        print(f"\nRaw Usage Metrics: {crew.usage_metrics}")

print_token_usage(result, crew, prompt_cost_per_million=0.2, completion_cost_per_million=1.5)
```
